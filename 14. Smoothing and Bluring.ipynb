```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# img = cv2.imread("opencv-logo.png")
# img = cv2.imread("face.jpg")
# img = cv2.imread("lena1.jpg")
img = cv2.imread("lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25
dst = cv2.filter2D(img, -1, kernel) # Homogenius filter: mean of neighbour
blur = cv2.blur(img, (5,5)) # blur function
gblur = cv2.GaussianBlur(img, (5,5), 0) # gausian filter different weight kernal, high in middle, low in side
median = cv2.medianBlur(img, 5) # median filter median of neighbour pixel. kernel except 1, use in black paper
bileteral = cv2.bilateralFilter(img, 9, 75,75) #biletral filter use for smooth edages and noise filter

titles = ['images', '2D Convolution','blur', 'gblur','median','bileteral']
images = [img, dst, blur, gblur, median, bileteral]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
