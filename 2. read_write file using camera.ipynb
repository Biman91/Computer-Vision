```python
import cv2
```

```python
cap = cv2.VideoCapture(0);
while True:
    ret, frame = cap.read() # frame capture frame, ret return True if frame available
    
    cv2.imshow("frame", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # wait for user input 0xFF for 64 bit
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# grayscale image
cap = cv2.VideoCapture(0);
while True:
    ret, frame = cap.read() # frame capture frame, ret return True if frame available
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # default BGR image, convert to gray
    cv2.imshow("frame", gray)
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # wait for user input 0xFF for 64 bit
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# dispaly image from video file
cap = cv2.VideoCapture(0);
fourcc = cv2.VideoWriter_fourcc(*'XVID') # 4 byte code, use for video codac, compressed format
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480)) # save the video, 20 frame per second, size

print(cap.isOpened())
while (cap.isOpened()):  #work if cap is exist
    ret, frame = cap.read() # frame capture frame, ret return True if frame available
    if ret==True: # if ret is TRUE then save file
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #get frame heigt
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        out.write(frame)

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # default BGR image, convert to gray
        cv2.imshow("frame", gray)

        if cv2.waitKey(1) & 0xFF == ord('q'): # wait for user input 0xFF for 64 bit
            break
    else:
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
