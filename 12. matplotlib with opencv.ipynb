```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lena.jpg", -1)

cv2.imshow("image", img)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img) 
# plt.xticks([]), plt.yticks([]) # to remove the ticks
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("gradient.png", 0)
_, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) 
_, th2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img, 150, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img, 150, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img, 150, 255, cv2.THRESH_TOZERO_INV)

titles = ["Image original", "Binary", "Binary Inverse", "Trunc", "Tozero", "Tozero Inverse"]
images = [img, th1, th2, th3, th4, th5]

for i in range(6):
    plt.subplot(2,3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
