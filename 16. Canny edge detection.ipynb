```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# canny edge detector: uses in multi-stage algorithm to detect a wide range of edges in images
# 1.Noise detection by using gausian filter
# 2. Gradiation Calculation to find intensiy of image
# 3. Non-maximum suppression to get rid off noise in edge
# 4. Double threshold to get potential edge
# 5. Edge tracking by hysterisis

img = cv2.imread("messi5.jpg", 0)

canny = cv2.Canny(img, 100, 200)

titles = ['images', 'canny']
images = [img, canny]
`
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
