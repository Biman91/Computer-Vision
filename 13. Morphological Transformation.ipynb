```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Morphological Transformation are simple opertion on image shaped
# perform on binary image
```

```python
img = cv2.imread("smarties.png", cv2.IMREAD_GRAYSCALE)

titles = ['image']
images =[img]

for i in range(1):
    plt.subplot(1,1, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

```python
img = cv2.imread("smarties.png", cv2.IMREAD_GRAYSCALE)
_, mask = cv2.threshold(img, 220,255, cv2.THRESH_BINARY_INV)

kernal = np.ones((5,5), np.uint8)
dialation = cv2.dilate(mask, kernal, iterations=2) #to remove the white dot in ball
erosion = cv2.erode(mask, kernal, iterations=1) # take care boundry of ball
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal) # followed erosion then dilation
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal) # followed dialation then erosion
mg = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal) # difference  erosion and dialation
th = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernal) # difference between image and opening

titles = ['image', 'mask', 'dilation', 'erosion', 'opening', 'closing', 'mg', 'th']
images =[img, mask, dialation, erosion, opening, closing, mg, th]

for i in range(8):
    plt.subplot(2,4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
